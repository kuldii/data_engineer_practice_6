```python
import pandas as pd
```

```python
filename_1 = '../assets/data/[1]game_logs.csv'
filename_2 = '../assets/data/[2]automotive.csv'
filename_3 = '../assets/data/[3]flights.csv'
filename_4 = '../assets/data/[4]dataset.csv'
```

```python
df_1 = pd.read_csv(filename_1, low_memory=False)
df_2 = pd.read_csv(filename_2, low_memory=False)
df_3 = pd.read_csv(filename_3, low_memory=False)
df_4 = pd.read_csv(filename_4, low_memory=False)
```

```python
file_memory_size_1 = df_1.memory_usage().sum() / (1024 * 1024)
file_memory_size_2 = df_2.memory_usage().sum() / (1024 * 1024)
file_memory_size_3 = df_3.memory_usage().sum() / (1024 * 1024)
file_memory_size_4 = df_4.memory_usage().sum() / (1024 * 1024)
```

```python
memory_size_loaded_1 = df_1.memory_usage(deep=True).sum() / (1024 * 1024)
memory_size_loaded_2 = df_2.memory_usage(deep=True).sum() / (1024 * 1024)
memory_size_loaded_3 = df_3.memory_usage(deep=True).sum() / (1024 * 1024)
memory_size_loaded_4 = df_4.memory_usage(deep=True).sum() / (1024 * 1024)
```

```python
column_stats_1 = []
for column in df_1.columns:
    col_memory = df_1[column].memory_usage(deep=True) / (1024 * 1024)
    col_dtype = df_1[column].dtype
    col_stats = {
        "column": column,
        "memory_usage": col_memory,
        "fraction_of_total": col_memory / memory_size_loaded_1,
        "dtype": str(col_dtype)
    }
    column_stats_1.append(col_stats)
```

```python
column_stats_2 = []
for column in df_2.columns:
    col_memory = df_2[column].memory_usage(deep=True) / (1024 * 1024)
    col_dtype = df_2[column].dtype
    col_stats = {
        "column": column,
        "memory_usage": col_memory,
        "fraction_of_total": col_memory / memory_size_loaded_2,
        "dtype": str(col_dtype)
    }
    column_stats_2.append(col_stats)
```

```python
column_stats_3 = []
for column in df_3.columns:
    col_memory = df_3[column].memory_usage(deep=True) / (1024 * 1024)
    col_dtype = df_3[column].dtype
    col_stats = {
        "column": column,
        "memory_usage": col_memory,
        "fraction_of_total": col_memory / memory_size_loaded_3,
        "dtype": str(col_dtype)
    }
    column_stats_3.append(col_stats)
```

```python
column_stats_4 = []
for column in df_4.columns:
    col_memory = df_4[column].memory_usage(deep=True) / (1024 * 1024)
    col_dtype = df_4[column].dtype
    col_stats = {
        "column": column,
        "memory_usage": col_memory,
        "fraction_of_total": col_memory / memory_size_loaded_4,
        "dtype": str(col_dtype)
    }
    column_stats_4.append(col_stats)
```

```python
result_json_1 = {"file_memory_size": file_memory_size_1,
               "memory_size_loaded": memory_size_loaded_1,
               "column_stats": column_stats_1}
result_json_2 = {"file_memory_size": file_memory_size_2,
               "memory_size_loaded": memory_size_loaded_2,
               "column_stats": column_stats_2}
result_json_3 = {"file_memory_size": file_memory_size_3,
               "memory_size_loaded": memory_size_loaded_3,
               "column_stats": column_stats_3}
result_json_4 = {"file_memory_size": file_memory_size_4,
               "memory_size_loaded": memory_size_loaded_4,
               "column_stats": column_stats_4}
```

```python
print(result_json_1)
print(result_json_2)
print(result_json_3)
print(result_json_4)
```
